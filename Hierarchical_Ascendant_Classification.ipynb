```python
#5- Changer l’échelle des données pour les ramener dans intervalle [0-1].

# Sélection uniquement des colonnes numériques
numerical_data = fromage.select_dtypes(include=['number'])
numerical_data.head(10)
```

```python
#Changer l’échelle des données pour les ramener dans intervalle [0-1].
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
data_normalized=scaler.fit_transform(numerical_data)
data_normalized
```

```python
#Pourquoi cette opération est-elle nécessaire ?
#La normalisation est nécessaire pour éviter que des variables à grande échelle dominent les variables à petite échelle .
```

```python
#6- Afficher le dendrogramme qui correspond à une classification hiérarchique ascendante.
#Utiliser la distance euclidienne et tester plusieurs critères d’agrégation :
#"indice de Ward", "le lien minimum", "le lien maximum" et "le lien moyen".
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

methods = ['ward', 'single', 'complete', 'average']
plt.figure(figsize=(13, 10))

for i, method in enumerate(methods, 1):
    plt.subplot(2, 2, i)
    Z = linkage(data_normalized, method=method, metric='euclidean')
    dendrogram(Z)
    plt.title(f'Dendrogramme - Méthode : {method}')

plt.tight_layout()
plt.show()
```

```python
#Analyser les dendrogrammes obtenus en identifiant, pour chacun des critères, le meilleur choix du nombre de classes k.
'''
1. Méthode de Ward  :

Cette méthode favorise la minimisation de la variance intra-classe.
On peut observer un découpage naturel en environ 3 ou 4 classes, en fonction de la hauteur où l'on coupe l'arbre.

2. Méthode du lien simple ("single linkage")  :

Cette méthode tend à créer des chaînes de points, ce qui peut générer des classes de tailles inégales.
Il semble difficile de distinguer un nombre optimal de classes, mais une séparation en 2 ou 3 classes pourrait être envisageable.

3. Méthode du lien complet ("complete linkage")  :

Cette approche donne des classes plus compactes.
Un seuil de coupe pertinent semble conduire à 3 classes, ce qui est visible par les différences de couleurs dans les branches.

Méthode du lien moyen ("average linkage")  :

Cette méthode donne un bon compromis entre les méthodes précédentes.
Une séparation en 3 classes semble raisonnable.'''
```

```python
#Méthode du Coude pour déterminer le nombre optimal de clusters

import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Calcul de l'inertie pour différents nombres de clusters
z = linkage(data_normalized, method='ward', metric='euclidean')
inertie = []
K = range(1, 11)  # Nombre de clusters de 1 à 10

for k in K:
    labels = fcluster(z, k, criterion='maxclust')
    # Calcul de la variance intra-cluster (inertie)
    variance = 0
    for cluster in np.unique(labels):
        points_cluster = data_normalized[labels == cluster]
        centroid = points_cluster.mean(axis=0)
        variance += np.sum((points_cluster - centroid) ** 2)
    inertie.append(variance)

# Visualisation de la méthode du coude
plt.figure(figsize=(8, 5))
plt.plot(K, inertie, 'bo-', markersize=8)
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie intra-cluster')
plt.title('Méthode du Coude pour déterminer le nombre optimal de clusters')
plt.grid(True)
plt.show()
```

```python
#7- Effectuer la catégorisation en utilisant la classification hiérarchique ascendante et visualiser les résultats obtenus et ce en utilisant :
#  Variables : seulement les 1er et 2ème attributs
#  Distance : euclidienne
#  Critères d’agrégation : "l’indice de Ward", "le lien maximum" et "le lien moyen".
#  Nombre de classes k : entre 2 et 5.

data= fromage.iloc[:, 1:3]
data_normalise=scaler.fit_transform(data)

methods = ['ward', 'complete', 'average']
k_values = [2, 3, 4, 5]

plt.figure(figsize=(20, 12))

plot_num = 1
for method in methods:
    Z = linkage(data_normalise, method=method, metric='euclidean')

    for k in k_values:
        clusters = fcluster(Z, k, criterion='maxclust')

        plt.subplot(3, 4, plot_num)
        plot_num += 1

        plt.scatter(data_normalise[:, 0], data_normalise[:, 1], c=clusters, cmap='rainbow')
        plt.title(f'{method.capitalize()} - k={k}', fontsize=10)
        plt.xlabel('Attr 1')
        plt.ylabel('Attr 2')

plt.tight_layout()
plt.show()
```
